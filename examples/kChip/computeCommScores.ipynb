```python
# contrast the actual and estimated assessments
# print(len(sims["M9_Alanine"]))
from json import dump
missing = {}
for med, models in sims.items():
    missing[med] = []
    for model in models:
        if model not in simConditions[med]:
            missing[med].append(model)

dump(missing, open("missing_pairs.json", 'w'), indent=3)
# print(len(missing))
# print(sims[:10])
# print(simConditions[:10])
```

```python
keyCombs = list(combinations(list(missing.keys()), 2))
print(len(keyCombs))
uniqueMissing = {}
numSame = 0
for c1, c2 in keyCombs:
    s1, s2 = set(missing[c1]), set(missing[c2])
    overlap = (s1-s2) | (s2-s1)
    if len(overlap) < 1:  numSame += 1
    uniqueMissing[c1+"_"+c2] = list(overlap)
    # print(c1, c2)
    # print(overlap)

print(numSame)
dump(uniqueMissing, open("uniqueMissing.json", 'w'), indent=3)
```
