```python
from modelseedpy.biochem import from_local

msdb = from_local("../../ModelSEEDDatabase/")
```

```python
cpdNames, cpdXRefs = {}, {}
for cpd in msdb.compounds:
    for name in cpd.names:
        cpdNames[name] = cpd.id
    cpdXRefs[cpd.id] = {k:list(v) if isinstance(v, set) else v for k,v in cpd.annotation.items()}
```

```python
cpd.annotation
```

```python
from json import dump

with open("compoundNames.json", "w") as jsonOut:
    dump(cpdNames, jsonOut, indent=4)

with open("compound_Xrefs.json", "w") as jsonOut:
    dump(cpdXRefs, jsonOut, indent=4)
```
