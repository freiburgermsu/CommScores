```python
models_to_simulate = ["L145", "L15", "L179", "L202", "L233", "L257", "L34", "L145", "L154", "L164", "L202", "L304", "L8"]
models_to_simulate = [mdl.replace("L", "leaf") for mdl in models_to_simulate]
models = [read_sbml_model(f"../data/models/vorholt models/{mdl}.xml") for mdl in models_to_simulate]
```

```python
%run ../commscores/GEMCompatibility.py
new_models = [GEMCompatibility.standardize(raw_model, exchanges=True) for raw_model in models]
```

```python
missing_mets = []
for mdl in new_models:
    for rxn in new_model.exchanges:
        for met in rxn.metabolites:
            if "cpd" in met.id: continue
            print(rxn.id, "\t\t", met.name)
            missing_mets.append(met)
```

```python
cpds = [cpd.id for cpd in msdb.compounds]
sorted_cpds = sorted(cpds, reverse=True)
print(sorted_cpds[0])
```

```python
msdb_cpd_template = {
        "abbreviation": "",
        "abstract_compound": None,
        "aliases": [
            ""
        ],
        "charge": 1,
        "comprised_of": None,
        "deltag": 10000000.0,
        "deltagerr": 10000000.0,
        "formula": "",
        "id": "cpd37000",
        "inchikey": "",
        "is_cofactor": 0,
        "is_core": 0,
        "is_obsolete": 0,
        "linked_compound": None,
        "mass": 10000000.0,
        "name": "",
        "notes": [
            ""
        ],
        "pka": "",
        "pkb": "",
        "smiles": "",
        "source": ""
    }
```

```python
missing_met = missing_mets[0]
print(missing_met.annotation, missing_met.charge, missing_met.compartment, missing_met.elements, missing_met.formula, missing_met.formula_weight, missing_met.y)
```

```python
populated_templates = []
for index, missing_met in enumerate(missing_mets):
    template_entry = msdb_cpd_template.copy()
    template_entry["aliases"] = missing_met.annotation
    template_entry["inchikey"] = missing_met.annotation["inchikey"]
    template_entry["mass"] = missing_met.formula_weight
    template_entry["notes"] = missing_met.notes
    template_entry["source"] = "CarveMe models"
    template_entry["name"] = missing_met.name
    template_entry["id"] = f"cpd{50000+index}"

    populated_templates.append()
```
