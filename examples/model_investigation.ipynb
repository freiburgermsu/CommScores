# parsing the MSDB

```python
from modelseedpy.biochem import from_local

msdb = from_local("../../ModelSEEDDatabase/")
```

```python
cpdNames, cpdXRefs = {}, {}
for cpd in msdb.compounds:
    for name in cpd.names:
        cpdNames[name] = cpd.id
    cpdXRefs[cpd.id] = {k:list(v) if isinstance(v, set) else v for k,v in cpd.annotation.items()}
```

```python
cpd.annotation
```

```python
from json import dump

with open("compoundNames.json", "w") as jsonOut:
    dump(cpdNames, jsonOut, indent=4)

with open("compound_Xrefs.json", "w") as jsonOut:
    dump(cpdXRefs, jsonOut, indent=4)
```

```python
from json import load

with open("../data/compoundNames.json", "r") as jsonIn:
    cpdNames = load(jsonIn)

with open("../data/compound_Xrefs.json", "r") as jsonIn:
    cpdXRefs = load(jsonIn)
```

```python
# def add_xrefs(xref, ID):
#     if isinstance(ID, list):  xref.extend(ID)
#     else:  xref.append(ID)
#     return xref
# display(cpdXRefs)
xrefs_dict = {}
for ID, xrefs in cpdXRefs.items():
    for xref, refID in xrefs.items():
        xref = xref.lower()
        refID = refID[0] if isinstance(refID, list) else refID
        xrefs_dict[xref] = {} if xref not in xrefs_dict else xrefs_dict[xref]
        xrefs_dict[xref][refID] = ID
xrefs_dict
```

```python
from json import dump

with open("MSDB_xRefs.json", "w") as jsonOut:
    dump({k.lower(): v for k,v in xrefs_dict.items()}, jsonOut, indent=3)
print(xrefs_dict.keys())
```
