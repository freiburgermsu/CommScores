# Original SBML2 models

```python
from pandas import read_csv
from glob import glob

simulated_models_df = read_csv(r"models_list.csv", names=["path", "modelName", "coreID", "flux?"])
simulated_models = [modelName.replace(" SEED model", "") for modelName in simulated_models_df["modelName"].to_numpy()]
display(simulated_models_df)
print(simulated_models)        
```
