```python
import pickle
import pathlib

import cobra
from cobra.io import read_sbml_model
import pandas as pd
```

```python
data_folder = pathlib.Path("../../data/processed/kchip")
model_folder = data_folder / "models"
media_folder = data_folder / "media"
```

```python
model_dict = {f"{model_file.stem.split('.')[0]}": read_sbml_model(model_file) for model_file in model_folder.glob("*.sbml")}
model_dict
```

```python
media_dict = {f"{media_file.stem.split('.')[0]}": list(pd.read_excel(media_file)["compounds"]) for media_file in media_folder.glob("*.xls")}
media_dict = {f"{k.removeprefix('M9_')}": v for k, v in media_dict.items() if not k.startswith("M9_No_Glucose")}
media_dict
```
