## Growth simulations of all models in all media conditions

```python
carbon_sources = dict(sorted({k: v[18:] for k, v in media_dict.items()}.items()))
carbon_sources
```

```python
predicted_growth_data = []
for model_name, model in model_dict.items():
    for media_name, media_list_names in media_dict.items():
        curr_model = model.copy()
        # update the medium
        curr_media = curr_model.medium
        media_list = [f"EX_{m}_e0" for m in media_list_names]
        for m in curr_media:
            if m in media_list:
                if m in carbon_sources[media_name]:
                    curr_media[m] = 10
                else:
                    curr_media[m] = 1000
            else:
                curr_media[m] = 0
        curr_model.medium = curr_media
        # simulate growth
        growth = curr_model.slim_optimize()
        data_item = {
            "model": model_name,
            "media": media_name,
            "growth": growth
        }
        predicted_growth_data.append(data_item)
```

```python
predicted_growth = pd.DataFrame(predicted_growth_data).pivot_table(index="model", columns="media", values="growth")
predicted_growth
```

```python
predicted_yield_df = predicted_growth.loc[row_labels, column_labels].apply(lambda row: row / row.max(), axis=1)
predicted_yield_df[predicted_yield_df < 0.1] = 0
predicted_yield_df
```

## Save experimental metabolic profile and predicted growth rate

```python
exp_yield = met_profile.reset_index().melt(id_vars="index", var_name="carbon", value_name="exp_yield")
pred_growth = predicted_growth.reset_index().melt(id_vars="model", var_name="carbon", value_name="pred_growth")
pred_growth.rename(columns={"model": "index"}, inplace=True)
yield_growth_df = exp_yield.merge(pred_growth, left_on=["index", "carbon"], right_on=["index", "carbon"], how="inner")
yield_growth_df
```

## Confusion matrix for predicted growth for each model in all media conditions

```python
predicted_growth_df = predicted_yield_df.copy()
predicted_growth_df[predicted_growth_df < 0.1] = 0
predicted_growth_df[predicted_growth_df > 0] = 1
predicted_growth_df = predicted_growth_df.astype(int)
predicted_growth_df
```

```python
# calculate tp, fp, tn, fn for every row for all columns between predicted_growth_df and growth_df
tp = ((predicted_growth_df > 0) & (growth_df > 0)).sum(axis=1)
fp = ((predicted_growth_df > 0) & (growth_df == 0)).sum(axis=1)
tn = ((predicted_growth_df == 0) & (growth_df == 0)).sum(axis=1)
fn = ((predicted_growth_df == 0) & (growth_df > 0)).sum(axis=1)
```

```python
confusion_df = pd.DataFrame({"TP": tp, "FP": fp, "TN": tn, "FN": fn})
confusion_df
```

```python
plot_data = confusion_df.reset_index().melt(id_vars="model", value_vars=["TP", "FP", "TN", "FN"], var_name="Type", value_name="Count")
g = sns.catplot(data=plot_data, x="model", y="Count", col="Type", kind="bar", aspect=2, col_wrap=2)
# change ylimit for each axis
for ax in g.axes:
    ax.set_ylim(0, predicted_growth_df.shape[1])
```
