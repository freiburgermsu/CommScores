```python
import pathlib
from cobra.io import read_sbml_model
```

```python
from commscores.gemcompatibility import GEMCompatibility
```

```python
model_path = pathlib.Path("../data/models/")
```

```python
raw_model = read_sbml_model(model_path / "vorholt_models/Leaf1.xml")
raw_model
```

```python
new_model = GEMCompatibility.standardize(raw_model, exchanges=True);
```

```python
raw_model.optimize()
```

```python
new_model.optimize()
```

```python
changed = 0
unchanged = 0
for rxn in new_model.exchanges:
    for met in rxn.metabolites:
        if "cpd" in met.id:
            changed += 1
        else:
            unchanged += 1
```

```python
print(f"Changed: {changed}")
print(f"Unchanged: {unchanged}")
```

```python
raw_sol = raw_model.optimize()
new_sol = new_model.optimize()
```

```python
print(raw_model.reactions.Growth.reaction)
```

```python
raw_model.reactions.Growth.reaction == new_model.reactions.Growth.reaction
```

```python
new_sol.fluxes["PTRCTA"]
```

```python
(raw_sol.fluxes - new_sol.fluxes).sort_values(ascending=False).head(10)
```
