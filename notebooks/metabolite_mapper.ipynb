```python
bigg_ms2 = {}
for cpd, alias in cpd_alias.items():
    if "BiGG" not in alias or cpd in list(bigg_ms2.values()):  continue
    for bigg in alias["BiGG"]:
        if bigg in bigg_ms2:  continue
        bigg_ms2[bigg] = cpd
print(bigg_ms2)

from json import dump
with open("BiGG->MSID.json", "w") as out:
    dump()
```

```python
import re
mappedIDs = {ID: bigg_ms2[re.sub("\[\w\]", "", ID)] for ID in IDs if re.sub("\[\w\]", "", ID) in bigg_ms2}
mappedIDs
```

```python
mapped_vitamins = []
for vita in vitamins["ID"].to_numpy():
    if vita in mappedIDs:  mapped_vitamins.append(mappedIDs[vita])
    else:  mapped_vitamins.append(vita)
    
mapped_nutrients = []
for nutrient in nutrients["ID"].to_numpy():
    if nutrient in mappedIDs:  mapped_nutrients.append(mappedIDs[nutrient])
    else:  mapped_nutrients.append(nutrient)
    
mapped_minimal_media = []
for cpd in minimal_media["compound"].to_numpy():
    if cpd in mappedIDs:  mapped_minimal_media.append(mappedIDs[cpd])
    else:  mapped_minimal_media.append(cpd)
    
print("vitamins:\t", mapped_vitamins)
print("\nnutrients:\t", mapped_nutrients)
print("\nmapped_minimal_media:\t", mapped_minimal_media)
```
